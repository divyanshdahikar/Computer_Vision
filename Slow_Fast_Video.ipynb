```python
cap = cv2.VideoCapture(r"C:\Users\divya\OneDrive\Pictures\Camera Roll\WIN_20251111_14_52_24_Pro.mp4")

while cap.isOpened():
    rat, frame = cap.read()
    if rat == True:
        frame = cv2.resize(frame, (500,600))
        cv2.imshow('SlowFast', frame)
        if cv2.waitKey(60) & 0xff == ord("q"):
            break
    else:
        break
cap.release()
cv2.destroyAllWindows()
```
