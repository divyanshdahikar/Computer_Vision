```python
import cv2
```

```python
img = cv2.imread(r"D:\coding\OpenCV\WSCubeTech Open CV\IMGS\download.jpeg")
img = cv2.resize(img, (500,600))

txt = cv2.putText(img,'Radhe',(115,80),cv2.FONT_HERSHEY_SCRIPT_SIMPLEX,3,(0,0,255),3,cv2.LINE_8,False)

txt = cv2.putText(img,'Radhe',(115,90),cv2.FONT_HERSHEY_SCRIPT_SIMPLEX,3,(0,0,255),3,cv2.LINE_8,True)

cv2.imshow('Text over img', txt)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
