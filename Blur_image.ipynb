```python
import cv2
import numpy as np
```

```python
img = cv2.imread(r"D:\coding\OpenCV\WSCubeTech Open CV\IMGS\download.jpeg")
res_img = cv2.resize(img, (500,600))

g = cv2.GaussianBlur(res_img,(15,15),0)
m = cv2.medianBlur(res_img, 25)
b = cv2.bilateralFilter(res_img, 15,300, 300)

h = np.hstack((g,m,b))
cv2.imshow('blur', h)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
