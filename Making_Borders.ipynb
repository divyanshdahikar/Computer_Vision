```python
import cv2
import numpy as np
```

```python
org_img = cv2.imread(r"D:\coding\OpenCV\WSCubeTech Open CV\IMGS\download.jpeg")

img = cv2.copyMakeBorder(org_img,20,20,20,20,cv2.BORDER_CONSTANT, None, value = 2)
img = cv2.resize(img, (200,300))

img1 = cv2.copyMakeBorder(org_img,20,20,20,20,cv2.BORDER_REFLECT, None, value = 2)
img1 = cv2.resize(img1, (200,300))

img2 = cv2.copyMakeBorder(img,20,20,20,20,cv2.BORDER_REFLECT_101, None, value = 2)
img2 = cv2.resize(img2, (200,300))

img3 = cv2.copyMakeBorder(img,20,20,20,20,cv2.BORDER_DEFAULT, None, value = 2)
img3 = cv2.resize(img3, (200,300))

img4 = cv2.copyMakeBorder(img,20,20,20,20,cv2.BORDER_REPLICATE, None, value = 2)
img4 = cv2.resize(img4, (200,300))

h = np.hstack((img, img1, img2, img3, img4))
cv2.imshow('Border', h)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
