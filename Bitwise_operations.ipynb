```python
import numpy as np
import cv2
```

```python
img1 = cv2.imread(r"D:\coding\OpenCV\WSCubeTech Open CV\IMGS\download.png")
img2 = cv2.imread(r"D:\coding\OpenCV\WSCubeTech Open CV\IMGS\Untitled.png")

img1 = cv2.resize(img1, (200,300))
img2 = cv2.resize(img2, (200,300))

new = cv2.bitwise_and(img1,img2)

h1 = np.hstack((img1,img2,new))

cv2.imshow('Bitwise and',h1)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
img1 = cv2.imread(r"D:\coding\OpenCV\WSCubeTech Open CV\IMGS\download.png")
img2 = cv2.imread(r"D:\coding\OpenCV\WSCubeTech Open CV\IMGS\Untitled.png")

img1 = cv2.resize(img1, (200,300))
img2 = cv2.resize(img2, (200,300))

new = cv2.bitwise_xor(img1,img2)

h3 = np.hstack((img1,img2,new))

cv2.imshow('Bitwise xor',h3)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
combine = np.vstack((h1,h2,h3))
cv2.imshow('Combine', combine)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
