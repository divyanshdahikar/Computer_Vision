```python
import cv2
```

```python
img = cv2.imread(r"C:\Users\divya\OneDrive\Pictures\IMG_20250828_202011.jpg")
img = cv2.resize(img, (500,500))
w,h = img.shape[0], img.shape[1]

Rotate_matrix = cv2.getRotationMatrix2D((w/2,h/2), -90, 1)
new_img = cv2.warpAffine(img, Rotate_matrix, (w,h))

cv2.imshow('rotated img', new_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
