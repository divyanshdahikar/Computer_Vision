```python
import cv2
```

```python
cap = cv2.VideoCapture(r"C:\Users\divya\CrossDevice\Infinix NOTE 40 5G\storage\DCIM\Camera\VID_20251029_104002.mp4")

while cap.isOpened():
    r, frame = cap.read()
    if r == True:
        frame = cv2.resize(frame, (500, 600))
        cv2.imshow('Video', frame)
        if cv2.waitKey(5) & 0xff == ord('p'):
            break
    else:
        break
cap.release()
cv2.destroyAllWindows()
```
