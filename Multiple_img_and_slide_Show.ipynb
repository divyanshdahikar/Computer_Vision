## resize, h_stack , vstack

```python
import cv2
import numpy as np
```

```python
img = cv2.imread(r"D:\coding\OpenCV\WSCubeTech Open CV\IMGS\download.jpeg")
resize_img = cv2.resize(img,(300,300))
# h = np.hstack((resize_img, resize_img))
v = np.vstack((resize_img, resize_img))
F = np.hstack((v,v))
# cv2.imshow('Hstack', h)
# cv2.imshow('Vstack', v)
cv2.imshow('Full Frame', F)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import os
```

```python
list_name = os.listdir(r"D:\coding\OpenCV\WSCubeTech Open CV\IMGS")
```

```python
for name in list_name:
    path = "D:\\coding\\OpenCV\\WSCubeTech Open CV\\IMGS"
    img_name = os.path.join(path, name)
    img = cv2.imread(img_name)
    if img is None:
        print(f"⚠️ Could not read {img_name}")
        continue
    img = cv2.resize(img, (500,600))
    cv2.imshow('slideShow', img)
    cv2.waitKey(1500)
cv2.destroyAllWindows()
```
