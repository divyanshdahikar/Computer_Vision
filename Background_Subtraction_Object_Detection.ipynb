```python
import cv2
```

```python
vid = cv2.VideoCapture(0)
sub_m = cv2.createBackgroundSubtractorMOG2()
while True:
    r, frame = vid.read()
    if r == True:
        frame = cv2.resize(frame, (700, 500))
        sub_v = sub_m.apply(frame)
        sub_v = cv2.bitwise_not(sub_v)
        cv2.imshow("Org Vid", frame)
        cv2.imshow("Subtraction Vid", sub_v)
        if cv2.waitKey(20) & 0xff == ord('q'):
            break
    else:
        break
vid.release()
cv2.destroyAllWindows()
```
