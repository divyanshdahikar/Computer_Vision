```python
import cv2
```

```python
cap = cv2.VideoCapture(r"D:\Screen Recording\Screen Recording 2025-11-16 181627.mp4")
c = 0
while True:
    r,frame = cap.read()
    if r == True:
        frame = cv2.resize(frame,(700,500))
        filename = ("D://coding//OpenCV//WSCubeTech Open CV//New folder//_org_img_"+str(c)+".png")
        cv2.imwrite(filename, frame)
        cv2.imshow('video', frame)
        c += 1
        if cv2.waitKey(30) & 0xff == ord('q'):
            break
    else:
        break
cap.release()
cv2.destroyAllWindows()
```
