```python
import cv2
```

```python
img = cv2.imread(r"D:\coding\OpenCV\WSCubeTech Open CV\IMGS\download.jpeg")
img = cv2.resize(img, (500,500))

img = cv2.circle(img, (263,165), 135, (0,0,255), 2, 4)

cv2.imshow('Circle Img', img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
img = cv2.imread(r"D:\coding\OpenCV\WSCubeTech Open CV\IMGS\images.jpeg")
img = cv2.resize(img, (500,500))

img = cv2.ellipse(img, (210,220), (110,200), 30, 0, 350, (0,0,255), 2, 4)

img = cv2.putText(img, 'Divyansh', (250,260), cv2.FONT_HERSHEY_COMPLEX_SMALL, 1, (255,0,0), 2, 4)
cv2.imshow('ellipse Img', img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
