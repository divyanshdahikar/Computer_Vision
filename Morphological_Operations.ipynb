```python
import numpy as np
import cv2
```

```python
img = cv2.imread(r"D:\coding\OpenCV\WSCubeTech Open CV\IMGS\images.jpeg",cv2.IMREAD_GRAYSCALE)
_,img = cv2.threshold(img, 127, 255, cv2.THRESH_BINARY)
img = cv2.resize(img, (400,300))

m = np.ones((10,10),np.int8)
er = cv2.erode(img, m, iterations = 1)
di = cv2.dilate(img, m, iterations = 1)

# cv2.imshow('Org img', img)
# cv2.imshow('Erosion img', er)
# cv2.imshow('Dilation img', di)

op = cv2.morphologyEx(img, cv2.MORPH_OPEN, m, iterations = 1)
CL = cv2.morphologyEx(img, cv2.MORPH_CLOSE, m, iterations = 1)
GR = cv2.morphologyEx(img, cv2.MORPH_GRADIENT, m, iterations = 1)
th = cv2.morphologyEx(img, cv2.MORPH_TOPHAT, m, iterations = 1)
bh = cv2.morphologyEx(img, cv2.MORPH_BLACKHAT, m, iterations = 1)

h = np.hstack((op,CL,GR,th,bh))

# cv2.imshow('Morph op', op)
cv2.imshow('Morph cl', CL)
# cv2.imshow('Morph gr', GR)
# cv2.imshow('Morph th', th)
# cv2.imshow('Morph bh', bh)


cv2. waitKey(0)
cv2.destroyAllWindows()
```
