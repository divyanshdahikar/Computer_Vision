```python
import cv2
```

```python
cap = cv2.VideoCapture(0)

while True:
    r, frame = cap.read()
    if r == True:
        frame = cv2.resize(frame, (500,600))
        cv2.imshow('Webcam', frame)
        if cv2.waitKey(1) & 0xff == ord('p'):
            break
    else:
        break
cap.release()
cv2.destroyAllWindows()
```
